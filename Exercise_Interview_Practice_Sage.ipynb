## 1. Plotting

```python
def f(x,y):
    return math.sin(y*y+x*x)/math.sqrt(x*x+y*y+.0001)
```

```python
plot3d(f,(-5,5),(-5,5), adaptive=True, color=rainbow(60, 'rgbtuple'), max_bend=.1, max_depth=15).show()
```

```python
polytopes.hypercube(4).plot().show()
```
